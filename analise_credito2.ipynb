**Exercício 7 – Transformação logarítmica dos pagamentos diferentes de zero**

Filtrando pagamentos diferentes de zero

```python
pay_nonzero = df[pay_features][df[pay_features] != 0]
```

O comando acima cria um novo conjunto de dados contendo apenas valores diferentes de zero.

Os valores iguais a zero são ignorados para evitar problemas matemáticos no cálculo do logaritmo.

Aplicando transformação logarítmica

```python
log_pay = pay_nonzero.apply(np.log10)
```

O método .apply() foi utilizado para aplicar a função np.log10() a todos os elementos do DataFrame.

A transformação logarítmica reduz a escala dos dados e facilita a visualização de distribuições muito assimétricas.

Plotando histogramas logarítmicos

```python
log_pay.hist(figsize=(12,8), bins=20)
plt.show()
```

Os histogramas das transformações logarítmicas mostram distribuições mais organizadas e menos distorcidas por valores extremos.

Após a aplicação do logaritmo, tornou-se mais fácil visualizar a distribuição dos pagamentos realizados pelos clientes.

Os dados passaram a apresentar comportamento mais próximo de uma distribuição normal, facilitando futuras análises estatísticas e modelos de machine learning.
